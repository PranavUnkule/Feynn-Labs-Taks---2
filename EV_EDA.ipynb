```python
import matplotlib
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cardetails = pd.read_csv('/content/EV_data.csv')
cardetails
```

```python
cardetails['Drive'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(cardetails['Drive'], palette="Set3")
```

```python
sns.set_theme(style="whitegrid")

g = sns.catplot( data=cardetails, kind="bar",x="Drive", y="NumberofSeats",ci="sd", palette="coolwarm_r", alpha=.6, height=6)
```

```python
cardetails
```

```python
plt.figure(figsize=(24,12))
# sns.countplot(df['Car_Name'])

ax = sns.countplot(x="Name", data=cardetails)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

```python
clean_data = pd.read_csv('/content/ElectricCarData_Clean.csv')
clean_data
```

```python
clean_data.info()
```

```python
clean_data['PlugType'].unique()
```

```python
plt.figure(figsize=(12,6))
sns.regplot(x="TopSpeed_KmH", y="AccelSec", data=clean_data);
```

```python
plt.figure(figsize=(12,6))
sns.regplot(x="TopSpeed_KmH", y="Range_Km", data=clean_data);
```

```python
clean_data['BodyStyle'].value_counts()
plt.figure(figsize=(12,6))
sns.countplot(clean_data['BodyStyle'], palette="Set3")
```

```python
sns.relplot(x="TopSpeed_KmH", y="AccelSec", hue="RapidCharge", style="RapidCharge",col="PowerTrain", col_wrap=5,height=3, aspect=.75, linewidth=2.5,kind="line", data=clean_data);
```

```python
sns.jointplot(data=clean_data,x="Range_Km", y="Efficiency_WhKm", hue="PlugType",kind="kde")
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(data=clean_data, x="PlugType", y="Efficiency_WhKm",split=True, inner="quart", linewidth=1,pallete="red")
```
